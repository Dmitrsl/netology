```python
#увеличим дефолтный размер графиков
from pylab import rcParams
rcParams['figure.figsize'] = 10,7
from mpl_toolkits.axes_grid1 import make_axes_locatable
# импортируем нужные библиотеки
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
import re
```

```python
files = os.listdir()
files = [file for file in files if re.match(r'\d{4}.csv', file)]
data = {}
for file in files:
    data[file[:4]] = pd.read_csv(file, engine='python')
data['2019'].head(3)
```

```python
data['2019']['Country or region'][data['2019']['Country or region'] == 'United States'] = 'United States of America';
```

```python
data['2019'].info()
```

```python
import geopandas as gpd
import geoplot
world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))
world = data['2019'].merge(world, left_on='Country or region', right_on='name')
world = gpd.GeoDataFrame(world)
```

```python
fig, ax = plt.subplots(1, 1, figsize=(20,10))
divider = make_axes_locatable(ax)
ax.set_title('Map of World Happiness Score', fontsize=48)
cax = divider.append_axes("right", size="7%", pad=0.1)
world.plot(column='Score',
           ax=ax,
           cax=cax,
           legend=True,
           legend_kwds={'label': "World Happiness Score"},
          );
```

Российская Федерация по уровню счастья ближе к среднемировым значениям

```python
df = data['2019'].merge(data['2016'][['Region', 'Country']],
                   left_on='Country or region', right_on='Country')
#sort_values(by='Score', ascending=False)
```

```python
group_mean = df.groupby('Region', as_index=False).mean().sort_values(by='Score', ascending=False)
group_mean
```

```python
# plt.barh(group_mean.Region, group_mean['Healthy life expectancy']);
fig, ax = plt.subplots(1, 1, figsize=(20,10))
ax.axes.set_title("GDP per capita",fontsize=50)
ax.set_xlabel("Region", fontsize=30)
ax.set_ylabel("GDP per capita", fontsize=30)
sns.boxplot(x="Region", y="GDP per capita",
                 data=df,
                 width=.75,
                 linewidth=2.5)
plt.xticks(rotation=45, fontsize=20);
plt.yticks(fontsize=20);
```

```python
В странах Западной Европы, Северной Америки, Австралии и Новой Зеландии
ВВП на душу населения довольно высок, что нельзя сказать о странах остального Мира.
При этом дисперсия распределения значений ВВП в этих регионах невелика,
относительно остальных стран, вероятно это связано с количеством стран в регионах
```

```python
fig, ax = plt.subplots(1, 1, figsize=(20,10))
ax.axes.set_title("Correlation Matrix",fontsize=50)
sns.heatmap(data['2019'].corr(), annot = True, fmt='.3g', square=True, cmap='coolwarm');
```

Наиболее ярко выражена связь между индексом счастья и ВВП на душу населеня, индексом социальной поддержки и коэффициентом продолжительности жизни.


Самый высокий коэффициет корреляции 0,835 между продолжительностью жизни и ВВП на душу населения.

```python
import plotly.express as px
fig = px.scatter(df, x="Score", y="GDP per capita", color="Region")
fig.show()
```

```python
fig = px.scatter(df,
                 x="Healthy life expectancy",
                 y="GDP per capita",
                 size="Score",
                 color="Region",
                 log_x=True,
                 size_max=60)
fig.show()
```

Данные будут неплохо кластеризоваться в пространстве ВВП-продолжительность жизни.

Зависимость нелинейная.

Интересно сгенерировать новый признак ВВП х продолжительность жизни, 

```python
feature = data['2019']["Healthy life expectancy"]*data['2019']["GDP per capita"]
feature.corr(data['2019']["Score"])
```

Получили 0.8323154422324124

Это самое высокое значене коэффициента корреляции для индекса счастья
