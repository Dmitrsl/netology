### 1) Как критики относятся к спортивным играм?

````python
# 1) Как критики относятся к спортивным играм?
games.Genre.value_counts()
````

````python
plt.figure(figsize=(16,8))
sns.boxplot(data=games, x='Genre', y='Critic_Score',
           # hue='ESRB_Rating' # 'Platform'
           )
````

H0  - средняя оценка Critic_Score спортивных игр выше, чем всех остальных

H1  - средняя оценка Critic_Score спортивных игр ниже, чем всех остальных

уровень значимости выбираeм 0.05

````python
games_sport = games.Critic_Score[games.Genre == 'Sports']
games_unsport = games.Critic_Score[games.Genre != 'Sports']
games_sport.mean(), games_unsport.mean(), games.Critic_Score.mean()
````

````python
stats.ttest_1samp(games_sport, 7.2045656634747255, nan_policy='omit')
````

statistic=1.549 > 0

Следовательно среднее значение оценки критиков спортивных игр больше неспортивных

pvalue/2=0.12184/2=0.06092 > 0.05

то есть нулевую гипотезу, о том что оценка критиков для спортивных игр выше, не отклоняем
Различия средних статистически не значимы

````python
stats.ttest_ind(games_sport, games_unsport, nan_policy='omit')
````

### 2) Критикам нравятся больше игры на PC или на PS4?

````python
games_pcps4 = games[(games.Platform == 'PC') | (games.Platform == 'PS4')]
games_pcps4.head()
````

Здесь различия в выборках статистически значимы

````python
games_pc.mean(), games_ps4.mean()
````

````python
stats.ttest_1samp(games_pc, 7.904587155963299, nan_policy='omit')
````

statistic=-7.433453350089602 < 0

````python
import re

from sklearn.feature_extraction.text import TfidfVectorizer
from nltk.corpus import stopwords
import nltk
nltk.download('stopwords')
nltk.download('wordnet')

from nltk.stem import WordNetLemmatizer

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, accuracy_score
````

````python
spam = pd.read_csv('spam.csv')
spam
````

````python
spam.Category.value_counts()
````

````python
def format_message(message: str, lang: str='english'):
    '''
    prepare text for classification
    message - text 
    lang - language for stopwords 
    '''
    
    stopwords_set = set(stopwords.words(lang))
    wordnet_lemmatizer = WordNetLemmatizer()
    
    message = re.sub('[\W_]+',' ', message.lower()).split()
    message = [word for word in message if word not in stopwords_set]
    message = [wordnet_lemmatizer.lemmatize(word) for word in message]
    
    return ' '.join(message)
````

````python
format_message(spam.Message[20])
````

````python
spam['format_message'] = spam.Message.apply(format_message)
spam
````

````python
tfidf = TfidfVectorizer(
    #stop_words='english',
    # use_idf=False,
    max_features=None,
    # smooth_idf=False,
    # analyzer='char_wb', #{‘word’, ‘char’, ‘char_wb’}
    sublinear_tf=True
)
tfidf_matrix = tfidf.fit_transform(spam.Message).toarray() #spam.format_message
# names = tfidf.get_feature_names()
# tfidf_matrix = pd.DataFrame(tfidf_matrix, columns=names)
# tfidf_matrix
````

````python
tfidf_matrix.shape
````

````python
X, y  = tfidf_matrix, spam.Category
````

````python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
````

````python
clf = LogisticRegression(random_state=42,
                         class_weight={'spam': 5},
                         solver='liblinear',
                        ).fit(X_train, y_train)
predict = clf.predict(X_test)
````

````python
accuracy_score(y_test, predict)
````

````python
confusion_matrix(y_test, predict)
````

Наилучшая точность 0.986244019138756 при преобразовании сырых сообщений, без предварительных манипуляций


Класс TfidfVectorizer включает в себя приведение к нижнему регистру, удаление стоп-слов, предобработку и тд

Значения по умолчанию довольно приемлемые

````python
spam['pred'] = clf.predict(tfidf_matrix)
spam
````
