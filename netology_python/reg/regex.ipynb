## Задание 1

Напишите функцию, которая принимает на вход строку и проверяет является ли она валидным транспортным номером (1 буква, 3 цифры, 2 буквы, 2-3 цифры). Обратите внимание, что не все буквы кириллического алфавита используются в транспортных номерах.

Если номер валиден, то функция должна возвращать отдельно номер и регион.

Примеры работы программы:  

1. `car_id = 'АВ222С96'`

Результат: 
`Номер АВ222С валиден. Регион: 96`

2. `car_id = 'АБ22ВВ193'`

Результат: 
`Номер не валиден`

````python
import re
def valid_avtonum(number : str = 'АВ222С96'):
    val = re.findall(r'^[АВЕКМНОРСТУХ]{2}\d{3}(?!=000)[АВЕКМНОРСТУХ]{1}\d{2,3}$', number)
    print(val)
    if val:
        print(f'Номер {number[:6]} валиден. Регион: {number[6:]}')
    else:
        print('Номер не валиден')      
            
valid_avtonum()
````

````python
valid_avtonum('ААА222С96')
````

Напишите функцию, которая будет удалять все последовательные повторы слов из заданной строки при помощи регулярных выражений.

Пример работы программы:  

`some_string = 'Напишите функцию функцию, которая будет будет будет будет удалять все все все все последовательные повторы слов из из из из заданной строки строки при помощи регулярных выражений'`

Результат: 
`Напишите функцию, которая будет удалять все последовательные повторы слов из заданной строки при помощи регулярных выражений.`

````python
some_string = 'Напишите функцию функцию, которая будет будет будет будет удалять все все все все последовательные повторы слов из из из из заданной строки строки при помощи регулярных выражений'
````

## Задание 3

Напишите функцию, которая будет возвращать акроним по переданной в нее строке со словами.

Примеры работы программы:  

1. `some_words = 'Информационные технологии'`

Результат: 
`ИТ`

2. `some_words = 'Near Field Communication'`

Результат: 
`NFC`

````python
some_words = 'Near Field Communication'
def get_acronim(some_words: str = some_words):
    print(re.sub(r'[\s+,]', r'', re.sub(r'(\w+)', lambda x: x[0][0], some_words)).upper())

get_acronim()
get_acronim("Напишите функцию, которая будет удалять все последовательные повторы слов из заданной строки при помощи регулярных выражений")
````

````python
some_words = 'Информационные технологии'
def get_acronim(some_words: str = some_words):
    for word in re.findall(r'(\w+)', some_words):
        print(re.findall(r'\b\w', word)[0].upper(), end='')
    print()

get_acronim()
get_acronim("Напишите функцию, которая будет удалять все последовательные повторы слов из заданной строки при помощи регулярных выражений")
````

## Задание 4

Напишите функцию, которая будет принимать на вход список email-адресов и выводить их распределение по доменным зонам.

Пример работы программы:  

`emails = ['test@gmail.com, xyz@test.in, test@ya.ru, xyz@mail.ru, xyz@ya.ru', xyz@gmail.com]`

Результат: 
```
gmail.com: 2
test.in: 1 
ya.ru: 2 
mail.ru: 1
```

````python
emails = ['test@gmail.com', 'xyz@test.in', 'test@ya.ru', 'xyz@mail.ru', 'xyz@ya.ru', 'xyz@gmail.com', 'test@gmail.com', 'xyz@test.in', 'test@ya.ru', 'xyz@mail.ru', 'xyz@ya.ru', 'xyz@gmail.com']
def domain_counter(emails: str = emails):
    str_emails = re.findall(r'(?<=@)\w+.\w{2,3}', ' '.join(emails))
    dict_emails = {domain: str_emails.count(domain) for domain in str_emails}
    [print(domain, dict_emails[domain]) for domain in dict_emails]
domain_counter()
````

## Задание 6 (необязательное)
Напишите функцию, которая будет проверять номер сотового телефона на валидность, если он валиден, то переводить его в формат:   
+7-xxx-xxx-xx-xx  
Постарайтесь предусмотреть как можно больше адекватных форматов изначального ввода номера.
Примеры работы программы:  

1. `phone = '+7 955 555-55-55'`

Результат: 
`+7-950-555-55-55`

2. `phone = '8(955)555-55-55'`

Результат: 
`+7-950-555-55-55`

3. `phone = '+7 955 555 55 55'`

Результат: 
`+7-950-555-55-55`

4. `phone = '7(955) 555-55-55'`

Результат: 
`+7-950-555-55-55`

5. `phone = '423-555-55-5555'`

Результат: 
`Номер не валиден`

6. `phone = '123-456-789'`

Результат: 
`Номер не валиден`

````python
phone =  '123-456-789'

def valid_cell(phone: str = phone):
    reg = r'(\+7|8).*?(\d{3}).*?(\d{3}).*?(\d{2}).*?(\d{2})'
    if re.findall(reg, phone):    
        return re.sub(reg, r'+7-\2-\3-\4-\5', phone)
    else:
        return 'Номер не валиден'

phones = ['+7 955 555-55-55', '+7-950-555-55-55', '8(955)555-55-55', '123-456-789',
         '+7 955 555 55 55', '+7-950-555-55-55', '7(955) 555-55-55', '423-555-55-5555']
for phone in phones:
    print(valid_cell(phone))
````
