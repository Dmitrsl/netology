### хотелось бы увидеть ваши комментарии по поводу расчета матрица корреляций двумя разными методами. Чем отличается результат? Какой метод лучше применять?

```python
peerson = hr.corr().abs().unstack().sort_values(ascending=False).drop_duplicates()[1:]
peerson
```

```python
spearman = hr.corr('spearman').abs().unstack().sort_values(ascending=False).drop_duplicates()[1:]
spearman
```

```python
peerson.index == spearman.index
```

### Если отранжировать матрицы корреляции по Пирсону и Спирману мы 
получим полное совпадение трех самых скореллированных пар признаков
в остальном значения корреляций примерно совпадают

То есть в данном случае корреляции рассцитанные двумя методами показывают одинаковый результат

Корреляция Пирсона чувствительна к нормольности распределения признаков, а также к выбросам,
Корреляция Спирмана - рассчитывается методом ранжирования признаков и не зависит от нормальности распределения

## 7. Проверить гипотезу, что сотрудники с высоким окладом проводят на работе больше времени, чем сотрудники с низким окладом

```python
high_salary = hr.average_montly_hours[hr.salary == 'high']
medium_salary = hr.average_montly_hours[hr.salary == 'medium']
low_salary = hr.average_montly_hours[hr.salary == 'low']
high_salary.mean(), medium_salary.mean(), low_salary.mean()
```

```python
sns.boxplot(data=[high_salary, medium_salary, low_salary])
```

Сотрудники с низким уровнем оплаты труда проводят на работе в среднем 200.99 часов

Н0 - сотрудники с высоким окладом проводят на работе меньше 200.99 часов

H1 - сотрудники с высоким окладом проводят на работе больше 200.99 часов

```python
stats.ttest_1samp(high_salary, 200.99)
```

Не отвергаем нулевую гипотезу


### Используем двухвыборочный критерий


Н0 - различие в количестве часов проводимых на работе сотрудниками получающих высокую и низкую зарабонтую плату статистически не значимо

H1 - различие в количестве часов проводимых на работе сотрудниками получающих высокую и низкую зарабонтую плату статистически значимо

```python
stats.ttest_ind(high_salary, low_salary, equal_var=False)
```

Принимаем уровень значимости 0.05

pvalue=0.44 > 0.05

Нулевую гипотезу о равенстве средних часов проводимых на работе сотрудниками получающими высокую и низкую зарабонтую плату отвергнуть не можем

Недостаточно информации, чтобы сделать вывод о различии средних в выборках

## 9. Разделить данные на тестовую и обучающую выборки Построить модель LDA, предсказывающую уволился ли сотрудник на основе имеющихся факторов (кроме department и salary) Оценить качество модели на тестовой выборкe

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, accuracy_score
```

```python
y = hr.left
X = hr.drop(['left', 'department', 'salary', 'salary_float'], 1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
lda = LinearDiscriminantAnalysis(solver = 'eigen',
                                 shrinkage='auto')
```

```python
lda.fit(X_train, y_train)
```

```python
predict = lda.predict(X_test)
```

```python
accuracy_score(y_test, predict), confusion_matrix(y_test, predict)
```

```python
sns.heatmap(confusion_matrix(y_test, predict), annot=True, center=0, cmap= 'coolwarm');
```

```python
# Логистическая регрессия

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
clf = LogisticRegression(
                         class_weight={0:1.75},
                         max_iter=200,
                         solver='liblinear',
                        ).fit(X_train, y_train)
predict = clf.predict(X_test)
accuracy_score(y_test, predict), confusion_matrix(y_test, predict)
```
