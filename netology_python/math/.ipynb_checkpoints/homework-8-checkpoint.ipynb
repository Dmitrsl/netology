### Задание 1

Получить интуицию, на что влияют параметры распределений. Построить по 5 функций плотностей с разными параметрами для каждого распределения: нормальное, экспоненциальное, Стьюдента на одном графике.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
import seaborn as sns
```

```python
plt.figure(figsize=(15,8))
for i in range(5):
    s = np.random.normal(i * 10, i + 1, 100_000)
    sns.distplot(s)
```

```python
plt.figure(figsize=(15,8))
for i in range(5):
    s = np.random.exponential(i + 5, 100_000)
    sns.distplot(s)
```

```python
plt.figure(figsize=(15,8))
for i in range(5):
    s = np.random.standard_t((i + 5) * (i + 5), 100_000)
    sns.distplot(s)
```

### Задание 2

Непрерывная случайная величина задана плотностью распределения:
$$ f(\xi) = C, \xi \in [0;5];f(\xi) = 0, \xi \notin [0;5]$$
Найдите C, математическое ожидание $\xi$ а также вероятность попадания $\xi$ в отрезок $[3.5;7]$, т.е. $P(3.5<\xi<7)$.

Плотность распределения это произодная функции распределения

$ f^\prime(\xi) = 1, \xi \in [0;5];$

$f^\prime(\xi) = 0, \xi \notin [0;5]$

Требуем, чтобы плотность распределения удовлетворяла условию:

$\int_{-\infty}^{+\infty} f(\xi)d\xi = \int_{-\infty}^{0} f^\prime(\xi)d\xi + \int_{0}^{5} f^\prime(\xi)d\xi + \int_{5}^{+\infty} f^\prime(\xi)d\xi =1 $

$\int_{0}^{5} f^\prime(\xi)d\xi = 1$

Не совсем понятно, что из этого должно следовать)

С другой стороны плотность вероятности равномерно распределенной величины

$C = \frac{1}{5-0} = \frac{1}{5}$

Математическое ожидание. 

$ М = \int_{-\infty}^{0} 0\xi d\xi + \int_{0}^{5} \frac{1}{5}\xi d\xi + \int_{5}^{+\infty} 0\xi d\xi = \frac{1}{5} \frac{25}{2} = 2.5$

можно считать:

$ М = \frac{5-0}{2} = 2.5$

вероятность попадания $\xi$ в отрезок $[3.5;7]$

$P=\int_{3.5}^{5} \frac{1}{5} d\xi + \int_{5}^{7} 0 d\xi$

$P=\frac{5}{5} - \frac{3.5}{5} = 0.3$

или

$P=5\frac{1}{5} - 3.5\frac{1}{5} = 0.3$
