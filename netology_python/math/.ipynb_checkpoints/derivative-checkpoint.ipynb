### Уровень 0
посчитайте значение 1-ой и 2-ой производной фунции

f(x) = x^5 + 4sin(2x) + cos(3x+3) в точке x = 1

```python
import numpy as np
from scipy.misc import derivative
from sympy import diff, symbols, cos, sin
from matplotlib import pyplot as plt
```

```python
def f(x):
    return x**5 + 4* np.sin(2*x) + np.cos(3*x +3)
print(derivative(f, 1.0, dx=1e-6 , n=1))
print(derivative(f, 1.0, dx=1e-6 , n=2))
```

```python
x = symbols('x')
dx1 = diff(x**5 + 4* sin(2*x) + cos(3*x +3))
dx1
```

```python
diff(dx1)
```

### Уровень 1**
Постройте график первой и второй производной функции

(sin(2x+1))^5

на отрезке [-5; 5]

```python
x = symbols('x')
d_1 = diff((sin(2*x+1))**5)
d_1
```

```python
diff(d_1)
```

```python
def f(x):
    return (np.sin(2*x+1))**5
x = np.linspace(-5,5,1000)
fx = f(x)
f1x = [derivative(f, xx, dx=1e-6, n=1) for xx in x]
f2x = [derivative(f, xx, dx=1e-6, n=2) for xx in x]
```

```python
plt.figure(figsize=(16, 7))
plt.plot(x, fx, label='function')
plt.plot(x, f1x, label='1st derivative')
plt.plot(x, f2x, label='2nd derivative')
plt.legend()
plt.show()
```

```python
# numpy
def d_fun(x, func):
    h = 1e-6
    return (func(x+h)-func(x-h))/(2*h)
def second_derivative(x):
    return 10 * (np.sin(2*x+1)**4) * np.cos(2*x +1)
f1x = d_fun(x, f)
f2x = d_fun(x, second_derivative)
```

```python
plt.figure(figsize=(16, 7))
plt.plot(x, fx, label='function')
plt.plot(x, f1x, label='1st derivative')
plt.plot(x, f2x, label='2nd derivative')
plt.legend()
plt.show()
```
