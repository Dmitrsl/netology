```python
import numpy as np
from scipy.optimize import differential_evolution
from matplotlib import pyplot as plt
import optuna
np.random.seed(42)
```

```python
def ackley(x):
    arg1 = -0.2 * np.sqrt(0.5 * (x[0] ** 2 + x[1] ** 2))
    arg2 = 0.5 * (np.cos(2. * np.pi * x[0]) + np.cos(2. * np.pi * x[1]))
    return -20. * np.exp(arg1) - np.exp(arg2) + 20. + np.e

bounds = [(-10, 10), (-10, 10)]
result = differential_evolution(ackley, bounds, seed=42)
result
```

```python
result.nit
```

```python
params = {'strategy': ['best1bin', 'best1exp', 'rand1exp', 'randtobest1exp', 'currenttobest1exp',
          'best2exp', 'rand2exp', 'randtobest1bin', 'currenttobest1bin', 'best2bin',
          'rand2bin', 'rand1bin',],
          'popsize': np.arange(10,21),
          'tol': np.linspace(1e-10, 0.1, 20),
          'mutation': np.linspace(0, 1.99, 20),
          'recombination': np.linspace(0, 1, 20),
          'updating': ['immediate', 'deferred'],
         }
```

```python
for param in params:
    list_iter = []
    for p in params[param]:
        list_iter.append(differential_evolution(ackley, bounds, **{param: p}, seed=42).nit)
    plt.figure(figsize=(16, 7))
    plt.plot(params[param], list_iter, label=param)
    plt.legend()
    plt.show()
```

```python
# параметры с минимальным количеством итераций
params = {'strategy': 'best1bin',
          'popsize': 11,
          'mutation': 0.0,
          'recombination': 1,
          'updating': 'immediate',
         }
result = differential_evolution(ackley, bounds, **params, seed=42)
result.nit
```

Функция Аскли имеет глобальный минимум равный нулю в точке (0, 0)

Попробую подобрать гиперпараметры, которые максимально приближают ее к глобальному 

минимуму, при этом требуют минимального количества итераций

```python
%%time
def objective(trial):
     
    params = {'strategy': trial.suggest_categorical('strategy', ['best1bin', 'best1exp',
                           'randtobest1bin', 'best2bin','rand1bin',]),
          'popsize': trial.suggest_int('popsize', 11, 19),
          'mutation': trial.suggest_uniform('mutation', 0, 0.75),
          'recombination': trial.suggest_uniform('recombination', 0.75, 1),
         }
    res = differential_evolution(ackley, bounds, **params, seed=42)
    loss = res.nit + res.fun * 1e10 + np.abs(res.x.sum()) *1e10
    return loss

study = optuna.create_study(direction='minimize')
study.optimize(objective, n_trials=100);
trial = study.best_trial

print('Accuracy: {}'.format(trial.value))
print("Best hyperparameters: {}".format(trial.params))
```
