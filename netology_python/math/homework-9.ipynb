```python
rv = sts.expon()
sample = rv.rvs(size=1000)
```

```python
x = np.linspace(0,2,100)
pdf = rv.pdf(x)
plt.hist(sample, density=True, bins=30, label='sample histogram')
plt.plot(x, pdf, label='theoretical pdf', alpha=0.5)
plt.legend()
plt.ylabel('fraction of samples, $f(x)$')
plt.xlabel('$x$')
plt.show()
```

```python
n1 = [2, 300, 10, 4, 15, 27, 156, 32, 3000]
samples_count = 1000
#массивы для сохранения средних по выборкам
sample = np.array([])
for number in range(samples_count):
    # на каждом цикле генерируем выборки разных размеров
    a1 = np.array([])
    for n in n1:
        sample1 = rv.rvs(size=n)
        a1 = np.append(a1, sample1.mean())
    
    sample  = np.append(sample, a1)
sample = sample.reshape((len(n1), -1))
```

```python
#строим графики
#подписи для графиков
for i, s in enumerate(sample):
    label_1 = 'sample of ' + str(n1[i]) + ' histogram'
    plt.hist(s, density=True, label=label_1, bins=50)
    plt.legend()
    plt.ylabel('samples means')
    plt.xlabel('$x$')
    plt.show();
```
