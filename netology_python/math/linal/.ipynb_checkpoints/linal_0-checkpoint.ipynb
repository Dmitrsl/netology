### Уровень 0:
Даны вектора x=(1,1), y=(2, 0) и z = (0,2)

Надо:

Изобразить вектора на экране

Изобразить точку x + y + z

Найти угол между векторами x и y


```python
v = np.array([[1,1], [2,0], [0,2]])
v
```

```python
plt.figure(figsize=(3, 3))
plt.arrow(0 , 0, v[0,0], v[0,1], head_width = 0.1,)
plt.arrow(0 , 0, v[1,0], v[1,1], head_width = 0.1,)
plt.arrow(0 , 0, v[2,0], v[2,1], head_width = 0.1,)
v_sum = v.sum(0)
plt.arrow(0, 0,  v_sum[0], v_sum[1],
          head_width = 0.2,
          width = 0.02,
          ls='-',
          color='red',
          alpha=0.5,)
plt.grid()
plt.show();
# v.sum(0) == v[0] + v[1] + v[2]
```

```python
# np.rad2deg(np.arccos(1 - pdist(v[:2], 'cosine')))
scalar = v[0].dot(v[1])
lin_0 = np.sqrt(v[0].dot(v[0]))
lin_1 = np.sqrt(v[1].dot(v[1]))
np.rad2deg(np.arccos(scalar / (lin_0 * lin_1)))
```

### Уровень 1:
Даны дома из Бостона

Выбрать любые n домов, образующих базис в пространстве признаков дома (все колонки, кроме medv)

Написать код разложения дома по указанному базису


```python
boston_data = pd.read_csv('boston_data.csv')
boston_data.head()
```

```python
boston_data = boston_data.iloc[:,:-1]
print(boston_data.shape)
boston_data.head()
```

```python
# определитель линейно независимых векторов отличен от нуля
np.linalg.det(boston_data[:13]), np.linalg.det(np.eye(13))
```

```python
det = 0
while det == 0:
    mask = np.random.choice(404, 13)
    det = np.linalg.det(boston_data.iloc[mask])
print(det)
basis = boston_data.iloc[mask]
basis
```

```python
def vec2basis(vector : np.array):
    return np.linalg.solve(basis, vector)
alphas = vec2basis(boston_data.iloc[0,:])
```

```python
boston_data.iloc[0,:], basis.dot(alphas)
```
