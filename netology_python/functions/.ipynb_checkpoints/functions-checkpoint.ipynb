Вы работаете секретарем и вам постоянно приходят различные документы. Вы должны быть очень внимательны, чтобы не потерять ни один документ. Каталог документов хранится в следующем виде:

```python
documents = [
    {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
    {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
    {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]
```

Перечень полок, на которых находятся документы хранится в следующем виде:

```python
directories = {
    '1': ['2207 876234', '11-2'],
    '2': ['10006'],
    '3': []
}
```

Необходимо реализовать пользовательские команды (внимание! это не название функций, которые должны быть выразительными, а команды, которые вводит пользователь, чтобы получить необходимый результат):
  
- p – people – команда, которая спросит номер документа и выведет имя человека, которому он принадлежит;  
- s – shelf – команда, которая спросит номер документа и выведет номер полки, на которой он находится  
*Правильно обработайте ситуации, когда пользователь будет вводить несуществующий документ*.
- l – list – команда, которая выведет список всех документов в формате   
**passport "2207 876234" "Василий Гупкин"**  
- as – add shelf – команда, которая спросит номер новой полки и добавит ее в перечень. *Предусмотрите случай, когда пользователь добавляет полку, которая уже существует*.
- d – delete – команда, которая спросит номер документа и удалит его из каталога и из перечня полок. *Предусмотрите сценарий, когда пользователь вводит несуществующий документ*;  
- m – move – команда, которая спросит номер документа и целевую полку и переместит его с текущей полки на целевую. *Корректно обработайте кейсы, когда пользователь пытается переместить несуществующий документ или переместить документ на несуществующую полку*;  
- a – add – команда, которая добавит новый документ в каталог и в перечень полок, спросив его номер, тип, имя владельца и номер полки, на котором он будет храниться. *Корректно обработайте ситуацию, когда пользователь будет пытаться добавить документ на несуществующую полку*.

```python
def people_command(number=None):
    if number is None:
        number = input('Введите номер документа: ')
    for document in documents:
        if document['number'] == number:
            return document['name']
            break
    return 'Нет совпадений по номеру документа'
   

def shelf_command(number=None):
    if number is None:
        number = input('Введите номер документа: ')
    for shelf in directories:
        if number in directories[shelf]:
            return shelf
            break
    return 'Нет совпадений по номеру документа'


def list_command():
    for document in documents:
        print(f'{document["type"]} "{document["number"]}" "{document["name"]}"')
    

def add_shelf():
    '''
    empty string for exit
    '''
    while True:
        number = input('Введите номер новой полки: ')
        if number == '':
            break
        elif number in directories:
            print('Такая полка уже существует')
        else:
            directories[number] = []
            print(f'Полка #{number} добавлена!')
    

def delete_command(number=None):
    if number is None:
        number = input('Введите номер документа: ')
    for i, document in enumerate(documents):
        if document['number'] == number:
            documents.pop(i)
            directories[shelf_command(number)].remove(number)
            return f' Документ №{number} удален!'
    return 'Нет совпадений по номеру документа'


def add_command(type_doc=None, number=None, name=None, shelf=None):
    if type_doc is None:
        type_doc = input('Введите тип документа: ')
    if name is None:
        name = input('Введите имя: ')
    if number is None:
        number = input('Введите номер документа: ')
    if shelf is None:
        shelf = input('Введите номер полки: ')
        
    if shelf in directories:
        documents.append({'type': type_doc, 'number': number, 'name': name})
        directories[shelf].append(number)
        print(f'Создан документ #{number}, помещен на полку "{shelf}"')
    else:
        print('Tакой полки не существует')

  
def move_command(number=None, shelf=None):
    if number is None:
        number = input('Введите номер документа: ')
    if shelf is None:
        shelf = input('Введите номер полки: ')
    if (number in sum(directories.values(), [])) and (shelf in directories):
        old_shelf = shelf_command(number)
        directories[shelf].append(number)
        directories[old_shelf].remove(number)
        print(f'Документ "{number}" перемещен на полку "{shelf}"!')
    else:
        print('Нет совпадений по номеру документа или такой полки не существует')
        
    
def main(documents : list = documents, directories : dict = directories):
    '''
        'exit', 'quit', '-q'  or empty string for exit
    '''
    while True:
        command = input("Введите команду: ")
        if command == 'exit' or command == 'quit' or command == '-q' or command == '':
            break
        elif command == 'p' or command == '-people':
            print(people_command())
        elif command == 's' or command == '-shelf':
            print(shelf_command())
        elif command == 'l' or command == '-list':
            list_command()
        elif command == 'as' or command == '-add shelf':
            add_shelf()
        elif command == 'd' or command == '-delete':
            print(delete_command())
        elif command == 'm' or command == '-move':
            move_command()
        elif command == 'a' or command == '-add':
            add_command()
        elif command == 'dir':
            print(directories)
        
    return 'The end'
```

```python
main()
```
