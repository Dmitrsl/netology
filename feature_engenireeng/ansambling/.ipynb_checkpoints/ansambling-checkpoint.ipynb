## Случайный лес

```python
# Подготовим функцию для кросс-валидации и обучим дерево решений
tree_params = {'criterion': 'mse',
              'splitter': 'best',
              'max_depth': 6,
              'min_samples_split': 30,
              'min_samples_leaf': 3,
              'min_weight_fraction_leaf': 0.0,
              'max_features': None,
              'min_impurity_decrease': 0.0,
              'min_impurity_split': None,
              'ccp_alpha': 0.0,}

def learn_regressor(X, y, estimator, polyfeatures=False, norm=False, random=42):
    score = []
    list_estimator = []
    if polyfeatures:
        poly = PolynomialFeatures()
        poly_X = poly.fit_transform(X)
        X = pd.DataFrame(poly_X, columns=poly.get_feature_names())

    for n, (train, test) in enumerate(KFold(n_splits=5, random_state=random, shuffle=True).split(y)):
        
        X_train = X.iloc[train]
        X_test = X.iloc[test]
        if norm:
            scaler = StandardScaler()
            X_train = scaler.fit_transform(X_train)
            X_test = scaler.transform(X_test)
        estimator = estimator.fit(X_train, y.iloc[train])        
        sc = estimator.score(X_test, y.iloc[test])
        score.append(sc)
        list_estimator.append(estimator)
        print(f'Fold {n}, R^2 coeff = {sc}')
    print(f'mean R^2 = {np.array(score).mean()}')
    
    return list_estimator
    

estimator = DecisionTreeRegressor(random_state=42,  **tree_params)
_ = learn_regressor(X_le, y, estimator, polyfeatures=False, norm=True)
estimator = DecisionTreeRegressor(random_state=42, **tree_params)
_ = learn_regressor(X_dummies, y, estimator, polyfeatures=False)
```

Деревья решений:

mean R^2 = 0.7689377308564989

```python
reg_rf = RandomForestRegressor( n_estimators=300,
                                max_depth=10,
                                max_features=0.75,
                                random_state=42,
                                n_jobs=-1)
lin_reg = LinearRegression()
print('RandomForest - label encoding')
list_rf_le = learn_regressor(X_le, y, reg_rf, polyfeatures=False, norm=False)
print('RandomForest - one hot encoding')
reg_rf = RandomForestRegressor(n_estimators=300, max_depth=10, max_features=0.75, random_state=42, n_jobs=-1)
list_rf_dummy = learn_regressor(X_dummies, y, reg_rf, polyfeatures=False)
print('LinearRegression')
list_lin_reg_le = learn_regressor(X_le, y, lin_reg)
print('LinearRegression dummy')
lin_reg = LinearRegression()
list_lin_reg_dummy = learn_regressor(X_dummies, y, lin_reg)
```

На данных с кодировкой label encoding, алгоритмы немного точнее,
поэтому в дальнейшем используем датасет  X_le для обучения

Полиноминальные признаки для алгоритмов на базе деневьев не намного точнее

Деревья решений:

mean R^2 = 0.7689377308564989

RandomForestRegressor:

mean R^2 = 0.8515624183641707 (polyfeatures 0.8517282920416841)

LinearRegression:

mean R^2 = 0.7559843204675979 (dummy 0.784576726493708)

##  Важность признаков

```python
imp = pd.DataFrame(index=X_le.columns,)
for i in range(5): 
    imp = imp.join(pd.DataFrame(data=list_rf_le[i].feature_importances_,
                   index=X_le.columns,
                   columns=[f'importances_fold_{i}']))
imp['importances'] = imp.mean(1)
imp = imp.sort_values(by='importances', ascending=False)
```

```python
# Посмотрим на первые 10
imp = imp[:10]
sns.barplot(x='importances', y=imp.index, data=imp);
```

## Стекинг
Будем использовать функцию 

```python
svr = make_pipeline(StandardScaler(), NuSVR(C=1e6, nu=0.85, gamma='auto'))

regressor = StackingRegressor(
    [   
        #('lr', LinearRegression()),  # linear regression
        ('svr', svr), # support vector regressor
        ('rf', reg_rf), # random forest
        ('dt', estimator) # desision tree
    ],
LinearRegression()) # linear regression
```

```python
list_regressors = learn_regressor(X_le, y, regressor, polyfeatures=False, norm=False)
```

mean R^2 = 0.8436287552867741

```python
pred = np.mean([reg.predict(X_le) for reg in list_regressors], 0)
plt.figure(figsize=(18,10))
sns.regplot(pred, y);
```

```python
plt.figure(figsize=(18,10))
sns.histplot((pred-y));
```

```python
np.corrcoef(pred, y)
```

Получили достаточно неплохую модель, с качеством R^2 = 0.84, что чуть ниже чем случайный лес, это объясняется тем, что для ансамблирования использовались слабые модели с низким качеством - линейная регрессия и дерево решений.
