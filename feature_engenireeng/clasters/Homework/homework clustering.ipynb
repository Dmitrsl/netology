## Bизуально разделить город на районы безотносительно голосов, подобрать оптимальное число кластеров

```python
k_inertia = []
ks = range(3,100)

for k in tqdm(ks):
    clf_kmeans = KMeans(n_clusters=k, max_iter=1000, n_init=30, algorithm='elkan')
    clusters_kmeans = clf_kmeans.fit_predict(geo_comment[['x_sc','y_sc']].values)
    k_inertia.append(clf_kmeans.inertia_)
```

```python
plt.figure(figsize=(15,8))
sns.lineplot(ks, k_inertia);
```

```python
diff = np.diff(k_inertia)
plt.figure(figsize=(15,8))
sns.lineplot(ks[1:], diff)
```

```python
diff_r = diff[1:] / diff[:-1]
plt.figure(figsize=(15,8))
sns.lineplot(ks[1:-1], diff_r)
k_opt = ks[np.argmin(diff_r)+1]
k_opt
```

Несмотря на эвристику, я считаю число кластеров 25 вполне оптимально разделяет множество на кластеры

```python
clf_kmeans = KMeans(n_clusters=25, max_iter=3000,)
geo_comment['clusters_all'] = clf_kmeans.fit_predict(geo_comment[['x_sc','y_sc']])
```

## Bыделить основные тематики комментариев жителей, можно использовать всю имеющуюся дополнительную информацию

```python
string = geo_comment.comment[0]
string
```

Удаляем из строк цифры, спецсимволы, дубликаты, а затем лемматизируем

Для ускорения работы берем случайные 10 слов комментария

```python
def prepare_string(string: str):
    '''
    delete digits, special characters and duplicates
    lemmatize 
    input: string
    output: string
    '''
    m = Mystem()
    string = re.sub(r'[^\w\s]+|[\d]', r'', string).strip()
    string = re.sub(r'\b([^\W\d_]+)(\s+\1)+\b', r'\1', string)
    string = ' '.join(np.random.choice(np.array(string.split()), size=10))
    # string = ''.join(m.lemmatize(string))
    
    return string
                    
prepare_string(geo_comment['comment'][0])
```

```python
%%time
geo_comment['comment_lemmas'] = geo_comment.comment.parallel_apply(prepare_string, 1)
geo_comment.head(1)
```

```python
# Здесь без лемматизации, тк очень долго работает
geo_comment['comment_lemmas'][0]
```

```python
corpus = geo_comment['comment_lemmas'].values[:10000]
corpus[:1]
```

Векторизуем, а затем пробуем кластеризовать

```python
vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(corpus)
X = X.toarray()
```

```python
X.shape
```

```python
pca = PCA(n_components=100)
X = pca.fit_transform(X)
sns.lineplot(pca.explained_variance_ratio_);
```

```python
X.shape
```

```python
geo_comment = geo_comment[:10000]
dbscan = DBSCAN(n_jobs=-1) # min_samples=250, 
geo_comment['clusters_comments'] = dbscan.fit_predict(X)
plt.figure(figsize=(15,8))
sns.scatterplot(data=geo_comment, x='x', y='y', hue='clusters_comments', palette="deep");
plt.show()
```

```python
# кластер улЖивописная
geo_comment[['comment_lemmas']][geo_comment['clusters_comments'] == 0]
```
