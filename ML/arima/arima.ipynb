```python
import pandas as pd
import numpy as np
import optuna
import matplotlib.pyplot as plt

import statsmodels.tsa.api as smt
from statsmodels.tsa.arima.model import ARIMA, ARIMAResults
from statsmodels.tsa.stattools import adfuller
from scipy.stats import boxcox

from utils import tsplot, test_stationarity
```

```python
car_sales = pd.read_csv("monthly-car-sales-in-quebec-1960.csv")
car_sales.tail()
```

```python
car_sales = car_sales['Count']
car_sales = boxcox(car_sales, 1)
car_sales = np.diff(car_sales, 2,)
tsplot(car_sales)
```

```python
def objective(trial):
    
#     method = trial.suggest_categorical("method", ['innovations_mle',
#                                                   'hannan_rissanen',
#                                                   'innovations',])
    i = trial.suggest_int("i", 0, 5, step=1)
    d = trial.suggest_int("d", 0, 5, step=1)
    j = trial.suggest_int("j", 0, 5, step=1)
    
    mdl = ARIMA(car_sales, order=(i,d,j), enforce_invertibility=True)
    mdl.fit()

    return mdl.k_ar

study = optuna.create_study()
study.optimize(objective, n_trials=50)

study.best_params  
```

```python
best_mdl = ARIMA(car_sales, order=(0,5,3))
res = best_mdl.fit()
tsplot(res.resid, lags=30)
```

```python
best_mdl = smt.ARIMA(car_sales, order=(1,2,5)).fit()
with plt.style.context('bmh'):
    plt.figure(figsize=(14,8))
    ax = plt.axes()
    best_mdl.plot_predict(2, len(car_sales)+20, ax=ax)
    plt.plot(car_sales, color='blue', label='Series')
    #plt.plot(res.resid, color='green', label='resid')
    plt.legend()
    plt.show()
```
