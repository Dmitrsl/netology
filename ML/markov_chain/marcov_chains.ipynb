```python
import numpy as np
import re
import os
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from pathlib import Path
from hmmlearn.hmm import GaussianHMM
from scipy.stats import boxcox
%matplotlib inline
sns.set()
```

```python
file_dir = Path('war_and_peace')
files = os.listdir(file_dir)
files
```

```python
all_text = ''
for file in files:
    with open(file_dir / file, 'r') as f:
        text = f.read().lower()
    all_text += re.sub(r'[^а-я\s\n]','', text) + " "
    
all_text = ' '.join(all_text.split())
```

```python
all_text[-500:]
```

### Mатрицa переходных вероятностей

```python
# буквы ё в тексте нет
RUSSIAN = 'абвгдежзийклмнопрстуфхцчшщъыьэюя '
POS = {l: i for i, l in enumerate(RUSSIAN)}
probabilities = np.zeros((len(RUSSIAN), len(RUSSIAN)))
for cp, cn in zip(all_text[:-1], all_text[1:]):
    probabilities[POS[cp], POS[cn]] += 1
    
probabilities = (probabilities / probabilities.sum(axis=1)[:, None])
df = pd.DataFrame(probabilities, index=(c for c in RUSSIAN), columns=(c for c in RUSSIAN))

plt.figure(figsize=(14,8))
sns.heatmap(df)
plt.show()
```

### Bероятность встретить строку из 4-х символов, которая начинается п, а кончается на р

выводим слова с вероятностью более 0.0001

```python
start = POS['п']
end = POS['р']
prior = 1. / len(RUSSIAN)
for c1 in RUSSIAN:
    for c2 in RUSSIAN:
        proba = prior * probabilities[start, POS[c1]] * probabilities[POS[c1],POS[c2]] * probabilities[POS[c2], end]
        if proba > 0.0001:
            print(''.join(['п', c1, c2, 'р']), proba)
```

### Cгенерируйте слово, начинающееся с буквы “Б” и имеющее длину 8-мь символов

```python
s = ['б']

length = 8
for i in range(length):
    index = POS[s[i]]
    next_char = np.random.choice(list(RUSSIAN), p=probabilities[index])
    s.append(next_char)

print (''.join(s))
```

Часто генерируются слова с пробелами, потому что в исходном тексте вероятность встретить слова длиной более 8-ми символов невелика. 

### Оцените ряд через HMM

```python
from hmmlearn.hmm import GaussianHMM
```

```python
os.listdir("../../arima")
```

```python
series = pd.read_csv("../../arima/international-airline-passengers.csv")[:-1]
series.columns = ["Month", "International airline passengers"]
```

```python
series = series["International airline passengers"]
```

```python
sns.lineplot(data=series)
```

Нормализуем дисперсию и дифференцируем значения ряда

```python

series = boxcox(series, 0)
series = np.diff(series, 1)
sns.lineplot(data=series);
```

```python
values = series.reshape(-1,1)
```

Визуально определяем на шесть уровней

```python
model = GaussianHMM(n_components=6,
                        covariance_type="diag",
                        n_iter=1000)
model.fit(values)
labels = model.predict(values)
means = np.zeros_like(values)
for i in range(model.n_components):
    means[labels == i] = model.means_[i]
    
plt.figure(figsize=(12, 6))
plt.plot(values)
plt.plot(means, linewidth=3)
```
