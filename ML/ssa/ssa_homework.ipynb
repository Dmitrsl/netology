```python
# Load the usual suspects:
import numpy as np
from numpy import pi
import matplotlib.pyplot as plt
import pandas as pd

from ssa import SSA

# Fiddle with figure settings here:
plt.rcParams['figure.figsize'] = (16, 5)
plt.rcParams['font.size'] = 14
plt.rcParams['image.cmap'] = 'plasma'
plt.rcParams['axes.linewidth'] = 2
# # Set the default colour cycle (in case someone changes it...)
# from cycler import cycler
# cols = plt.get_cmap('tab10').colors
# plt.rcParams['axes.prop_cycle'] = cycler(color=cols)

# # A simple little 2D matrix plotter, excluding x and y labels.
# def plot_2d(m, title=""):
#     plt.imshow(m)
#     plt.xticks([])
#     plt.yticks([])
#     plt.title(title)
```

```python
F = pd.read_csv('international-airline-passengers.csv').iloc[:-1,1].values
```

```python
plt.plot(F);
plt.title("international airline passengers");
```

```python
for components in range(6, 33, 3):

    F_ssa = SSA(F, components)
    F_ssa.components_to_df().plot()
    F_ssa.orig_TS.plot(alpha=0.3).legend(fontsize=10, ncol=2)
    plt.xlabel("$t$")
    plt.ylabel(r"$\tilde{F}_i(t)$")
    plt.title(f"$L={components}$ for international airline passengers");
    plt.show();
    F_ssa.plot_wcorr()
    plt.title(f"W-Correlation for international airline passengers, $L={components}$");
    plt.show();
```

### L = 15

При L > 10 начинет ярко проявляться трендовая составлющая
При L > 30 длина окна оказывает все меньшее влияние

w-корреляционная матрица для L = 15 представляет собой двухблочную структуру, с $\tilde{F}_0, \ldots, \tilde{F}_5$ в первом блоке и $\tilde{F}_6, \ldots, \tilde{F}_{14}$ во втором. 

Сгруппируем компоненты следующим образом:
\begin{align*}
    \tilde{F}^{(0)} & = \tilde{F}_0 \\
    \tilde{F}^{(1)} & = \tilde{F}_1 + \tilde{F}_2 + \tilde{F}_3 + \tilde{F}_4 + \tilde{F}_5 \\
    \tilde{F}^{(3)} & = \tilde{F}_6 + \ldots + \tilde{F}_{14} \\
\end{align*}



```python
F_ssa = SSA(F, 15)
F_ssa.reconstruct(0).plot()
F_ssa.reconstruct([1,2,3,4,5]).plot()
F_ssa.reconstruct(slice(6,40)).plot(alpha=0.7)
plt.title("Component Groupings for international airline passengers, $L=15$")
plt.xlabel("$t$")
plt.ylabel(r"$\tilde{F}_i(t)$")
plt.legend([r"$\tilde{{F}}^{{({0})}}$".format(i) for i in range(4)]);
```

Компонент $\tilde{F}(0)$ трендовая составляющая

Компонент $\tilde{F}(1)$ периодическая составляющая

Компонент $\tilde{F}(2)$ - шум

### L = 27

При L = 27 стуктура становится похожей на трехблочную

Сгруппируем компоненты следующим образом:
\begin{align*}
    \tilde{F}^{(0)} & = \tilde{F}_0 \\
    \tilde{F}^{(1)} & = \tilde{F}_1 + \tilde{F}_2 \\
    \tilde{F}^{(2)} & = \tilde{F}_3 + \tilde{F}_4 \\
    \tilde{F}^{(3)} & = \tilde{F}_5 \\
    \tilde{F}^{(5)} & = \tilde{F}_6 +  \tilde{F}_{7} \\
    \tilde{F}^{(6)} & = \tilde{F}_8 +  \tilde{F}_{9} \\
    \tilde{F}^{(7)} & = \tilde{F}_{10} + \ldots + \tilde{F}_{27} \\
\end{align*}


```python
F_ssa = SSA(F, 27)
F_ssa.reconstruct(0).plot()
F_ssa.reconstruct([1,2]).plot()
F_ssa.reconstruct([3,4]).plot()
F_ssa.reconstruct(5).plot()
F_ssa.reconstruct([6,7]).plot()
F_ssa.reconstruct([8,9]).plot()
F_ssa.reconstruct(slice(10,27)).plot(alpha=0.3)
plt.title("Component Groupings for international airline passengers, $L=27$")
plt.xlabel("$t$")
plt.ylabel(r"$\tilde{F}_i(t)$")
plt.legend([r"$\tilde{{F}}^{{({0})}}$".format(i) for i in range(7)]);
```

Компонент $\tilde{F}(0)$ трендовая составляющая

Компоненты $\tilde{F}(1) \ldots \tilde{F}(10)$  периодические составляющие

Компонент $\tilde{F}(5)$ - низкочастотная периодическая составляющая (сезонность)

Компонент $\tilde{F}(10) \ldots \tilde{F}(27)$  - шум

Сумма первых девяти компонет вполне отражает исходный верменной ряд.

```python
n_components = 9
plt.plot(F);
F_ssa.reconstruct(slice(0,n_components)).plot()
F_ssa.reconstruct(slice(n_components,27)).plot(alpha=0.3)
plt.title("Component Groupings for international airline passengers, $L=27$")
plt.xlabel("$t$")
plt.ylabel(r"$\tilde{F}_i(t)$")
plt.legend(['Исходный временной ряд', 'Сумма периодических компонент','Шум']);
```
